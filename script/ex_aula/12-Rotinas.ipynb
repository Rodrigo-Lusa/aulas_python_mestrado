```python
def contar_inteiros(objeto: list) -> int:
    contador = 0
    for i in range(len(objeto)):
        if type(objeto[i]) == int:
            contador += 1
    return i

lista = [1, 3, 5]

print(contar_inteiros(lista))
print(contar_inteiros([1, 4, 6, "eita", 8]))
print(contar_inteiros(objeto=[1, 4, 6, "eita", 8]))
```
