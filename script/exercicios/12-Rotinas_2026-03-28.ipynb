## 04 – Escreva o programa em Python:

```python
import numpy as np

array= np.random.normal(10, 3, 5)
array


def maior_que_x(objeto: array, valor_de_x: int) -> None:
    '''
    Dado um array e um número inteiro (x).
    O programa printa na tela os valores contidos no array maior que o número e depois o total de valores acima de x.
    '''
    numero = 0
    for i in objeto:
        if i > valor_de_x:
            print(i)
            numero += 1
    print(f"Número de valores maior que {valor_de_x}: {numero}")

maior_que_x(array, 10)
```

## 05 – Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def divisao(a: int | float, b: int | float) -> float:
    if type(a) not in [int, float] or type(b) not in [int, float]:
        return print("O parametro deve ser um número inteiro ou decimal.")
    if b == 0:
        print("Nao pode dividir por zero.")
    else:
        return a / b
```

```python
divisao(10, 2)
```

```python
divisao(10,0)
```

```python
divisao('a', 'b')
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def divisao_composta(a: list | tuple, b: list | tuple):
    if type(a) not in [list, tuple] or type(b) not in [list, tuple]:
        print("Os parâmetros devem ser listas ou tuplas.")
        return
    
    if len(a) != len(b):
        print("As estruturas devem ter o mesmo tamanho.")
        return
    
    resultado = [] 
    
    for i in range(len(a)):
        x = a[i]
        y = b[i]
        
        if type(x) not in [int, float] or type(y) not in [int, float]:
            print(f"Valor inválido. i = {i}.")
            return
        
        if y == 0:
            print("Não é possível fazer divisão por 0.")
            return
        
        resultado.append(x / y)
    
    print(resultado)
```

```python
divisao_composta([2, 4, 6], [1, 2, 3])
```

```python
divisao_composta((2, 4, 6), (1, 2, 3))
```

```python
divisao_composta((2, 4, 6), (1, 0, 3))
```

```python
divisao_composta((2, 4, 6), (1, 1, 3, 4))
```

```python
divisao_composta((2, 4, 6), 'a')
```
