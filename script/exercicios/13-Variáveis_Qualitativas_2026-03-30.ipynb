```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("df_gene_info.csv", index_col='Unnamed: 0')
print(df)
```

```python
# Tabela de frequência do número de caracteres no nome dos genes?
freq = df['len'].value_counts()

# Gráfico de barras deste dado?
grafico1 = freq.plot.bar()

grafico1.set_title("Frequência do numero de caracteres nos nomes dos genes")
grafico1.set_xlabel("Número de caracteres do gene")
grafico1.set_ylabel("Frequência")
plt.xticks(rotation=0)
```

## 04 – Descreva o racional da sua resposta:

R.: A variável analisada corresponde ao número de caracteres nos nomes dos genes, sendo uma variável quantitativa discreta. Como existe uma ordem natural entre os valores, foi necessário organizar os dados de forma crescente utilizando o método sort_index(). Em seguida, foi construída uma tabela de frequência absoluta por meio da função value_counts(), permitindo identificar a distribuição dos tamanhos dos nomes dos genes.
Por fim, os dados foram representados graficamente por um gráfico de barras, facilitando a visualização da frequência de cada categoria.

## 06 – Escreva o programa em Python:

```python
# Ordenar o df pelo index
df_sorted = df.sort_index()
freq_absoluta = df_sorted['len'].value_counts()
freq_rel = df_sorted['len'].value_counts(normalize=True)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 linha, 2 colunas

# Gráfico 1 - Frequência absoluta
freq_absoluta.plot.bar(ax=axes[0])
axes[0].set_title('Frequência Absoluta')
axes[0].set_xlabel('Nº de caracteres')
axes[0].set_ylabel('Frequência')

# Gráfico 2 - Frequência relativa
freq_rel.plot.bar(ax=axes[1])
axes[1].set_title('Frequência Relativa')
axes[1].set_xlabel('Nº de caracteres')
axes[1].set_ylabel('Proporção')

plt.tight_layout()
plt.show()
```

```python
freq_acumulada = freq_absoluta.cumsum()

acum = freq_acumulada.plot.bar()
acum.set_xlabel("Número de caracteres do gene")
acum.set_ylabel("Frequência acumulada")
plt.xticks(rotation=0)
```
