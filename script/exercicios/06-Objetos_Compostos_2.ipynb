## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
florida = pd.read_csv('../data/data.csv')
florida.head()
```

```python
linhas = florida.shape[0]
```

```python
classes_colunas = list(florida.dtypes)
```

```python
#florida.dtypes.value_counts()
n_elementos_classes = florida.dtypes.value_counts().to_list()
n_elementos_classes
```

```python
florida_stats = florida[['policyID', 'eq_site_limit', 'tiv_2012']].describe()

florida_stats
```

```python
mean = florida_stats.loc['mean'].to_list()
std = florida_stats.loc['std'].to_list()
terceiro_quartil = florida_stats.loc['75%'].to_list()

print(mean)
print(std)
print(terceiro_quartil)
```

```python
std = florida_stats.loc['std'].to_list()
std
```

```python
dict_florida = {
    'n_entradas': linhas,
    'classes_colunas': classes_colunas,
    'n_elementos_classes': n_elementos_classes,
    'media': mean,
    'desv': std,
    'terceiro_quartil': terceiro_quartil,
    }

dict_florida
```
