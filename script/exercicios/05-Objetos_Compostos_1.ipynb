## 09 – Instancie cinco variáveis da classe lista. Cada uma delas será referente à um dos valores contidos na lista instanciada no exercício 4 (Nomes dos genes). Para isso você deve utilizar a indexação de lista. Retorne cada um dos objetos:

```python
oxa_23 = [len(lista_de_genes[0]), lista_de_genes[0][0:2], lista_de_genes[0].count("a"), 'hu' in lista_de_genes[0]]
kpc = [len(lista_de_genes[1]), lista_de_genes[1][0:2], lista_de_genes[1].count("a"), 'hu' in lista_de_genes[1]]
vim = [len(lista_de_genes[2]), lista_de_genes[2][0:2], lista_de_genes[2].count("a"), 'hu' in lista_de_genes[2]]
ndm = [len(lista_de_genes[3]), lista_de_genes[3][0:2], lista_de_genes[3].count("a"), 'hu' in lista_de_genes[3]]
tet = [len(lista_de_genes[4]), lista_de_genes[4][0:2], lista_de_genes[4].count("a"), 'hu' in lista_de_genes[4]]

print(oxa_23)
print(kpc)
print(vim)
print(ndm)
print(tet)
```
