## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: Etapa 1, criar um loop que itere 5000 valores inteiros, iniciando em 0, e adicionando cada iteração em uma lista. Etapa 2: usar outro loop para ver cada um dos itens na lista. Se for par (o resto da divisão por 2 ser 0), vai ser adicionado na lista de numeros pares; se não, no dos ímpares.

Para saber o numero total vai ser usado o método len().

## 02 – Desenhe o fluxograma da resposta:

R.: <img src="./img/ex2_bloco11.png" width="400" />

## 03 – Escreva o programa em Python:

```python
def par_ou_impar(num: int = 5000) -> None:
    lista = []
    par = []
    impar = []

    # Cria a lista de 0 a 'num' (default = 5000)
    for i in range(num):
        lista.append(i)
    
    # Verifica se é impar ou par
    for i in lista:
        if i % 2 == 0:
            par.append(i)
        else:
            impar.append(i)

    print(f"Números pares: {len(par)}")
    print(f"Números pares: {len(impar)}")

par_ou_impar()
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
def par_ou_impar(num: int = 5000) -> None:
    lista = []
    par = []
    impar = []

    # Cria a lista de 0 a 'num' (default = 5000)
    while len(lista) < num:
        lista.append(len(lista))
    
    # Verifica se é impar ou par
    for i in lista:
        if i % 2 == 0:
            par.append(i)
        else:
            impar.append(i)

    print(f"Números pares: {len(par)}")
    print(f"Números pares: {len(impar)}")

par_ou_impar()
```

## 07 – Escreva o programa em Python:

```python
import pandas as pd

df = pd.read_csv("df_gene_info.csv", index_col='Unnamed: 0')
print(df)
```

```python
# Dos números de letras nos genes, quais são pares e quantos são impares?

def len_info(tabela) -> None:
    for linha in tabela.itertuples():
        if linha[1] % 2 == 0:
            print(f"O gene {linha[0]} tem um número par ({linha[1]}) de caracteres.")
        else:
            print(f"O gene {linha[0]} tem um número ímpar ({linha[1]}) de caracteres.")
            
len_info(df)
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
def len_info(tabela) -> None:
    for i in range(len(tabela)):
        linha = tabela.iloc[i]
        gene = tabela.index[i]
        valor = linha.iloc[0] 

        if valor % 2 == 0:
            print(f"O gene {gene} tem um número par ({valor}) de caracteres.")
        else:
            print(f"O gene {gene} tem um número ímpar ({valor}) de caracteres.")

len_info(df)
```
