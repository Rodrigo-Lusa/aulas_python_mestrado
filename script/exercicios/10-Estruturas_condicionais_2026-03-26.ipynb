## 04 – Escreva o programa em Python:

```python
import numpy as np

array= np.random.normal(10, 3, 5)
array
```

```python
def maior_que_x(objeto: array, valor_de_x: int) -> None:
    '''
    Dado um array e um número inteiro (x).
    O programa printa na tela os valores contidos no array maior que o número e depois o total de valores acima de x.
    '''
    numero = 0
    for i in objeto:
        if i > valor_de_x:
            print(i)
            numero += 1
    print(f"Número de valores maior que {valor_de_x}: {numero}")

maior_que_x(array, 10)
```

## 05 – Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

```python
# Sim, a única diferença é que na hora de chamar a função, o valor do segundo parametro (valor_de_x) deve ser mudado para 20

maior_que_x(array, 20)
```

R.: O que eu pretendo fazer é definir uma função que aceita como parametro uma tabela (DataFrame).
Vou achar um método de iterar pelas linhas desse df com um loop. Para cada uma das linhas, se o valor presente no segundo item (linha[1]) for >=2 e <= 4, vou printar suas informações (a linha inteira), se for menor do que 2, vou printar aqueles valores onde forem maior do que 4. ## 09 – Escreva o programa em Python:

```python
import pandas as pd

df = pd.read_csv("df_gene_info.csv", index_col='Unnamed: 0')
df
```

```python
def len_info(tabela) -> None:
    for linha in tabela.itertuples():
        
        # Caso o nome tenha entre 2 e 4 caracteres inclusive, posso saber as informações do gene
        if linha[1] >= 2 and linha[1] <= 4:
            print(f"{linha[0]} ({linha[1]} caracteres) info:")
            print(linha)
            print()

        # Se tiver menos que 2
        elif linha[1] < 2:
            print(f"O gene {linha[0]} tem menos de 2 caracteres.")
            # Saber os valores onde os genes tem mais de 4 caracteres
            if linha[1] > 4:
                print(f"Mas vou mostrar os que tem mais de 4 haha:")
                print(linha)
                print()

        # Se tiver mais de 4, informar quem
        elif linha[1] > 4:
            print(f"O gene {linha[0]} tem mais de 4 caracteres.")
            print()
        

len_info(df)
```

R.:  Sim, pois a função len_info() tem um parametro de entrada, que é um dataFrame. Assim, independente de seu tamanho, a função irá iterar sobre ele e printar as informações de acordo com as condições estabelecidas.
