## 01 – Dados as operações abaixo, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

p2 = 2.0 + 3.0 / 5.0 != 10
q2 = (p + 10.2) - 5.0 * 2.0 == 2.9
r2 = 10 <= p / q + 33.8 > 100
s2 = 5 == p / (q + r)
t2 = 33 != s * 2

print("Antes Depois")
print(p, p2)
print(q, q2)
print(r, r2)
print(s, s2)
print(t, t2)
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
p = ('a',1,True,'b',0,False) or ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] or ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] or ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] or ['a',True,1]
t = ['a',1,True,'b',0,False] or 1
u = ['a',True,1] <= ['a',True,1] or ['a',1,True,'b',0,False]

print(p)
print(q)
print(r)
print(s)
print(t)
print(u)
```

```python
p2 = not ('a',1,True,'b',0,False) or ('a',True,1,'b',False,0)
q2 = not ['a',1,True,'b',0,False] or ('a',True,1,'b',False,0)
r2 = not ['a',1,True,'b',0,False] or ['a',True,1,'b',False,0]
s2 = not ['a',1,True,'b',0,False] or ['a',True,1]
t2 = not ['a',1,True,'b',0,False] or 1
u2 = not ['a',True,1] <= ['a',True,1] or ['a',1,True,'b',0,False]

print(p, p2)
print(q, q2)
print(r, r2)
print(s, s2)
print(t, t2)
print(u, u2)
```
