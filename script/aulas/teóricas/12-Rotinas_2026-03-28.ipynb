```python
# Instanciando uma variável do tipo string
m = "Python é colorido e divertido"
print(m)
```

```python
# Verificando o número de elementos da string
len(m)
```

```python
# Usando um loop for para iterar sobre cada caractere da string
for id, item in enumerate(m):
    print(item, end="")
```

```python
# Arredondando o resultado da divisão do número de caracteres por 225
round(len(m)/225, 2)
```

```python
# Multiplicando a string por um número inteiro para repetir seu conteúdo
tamanho = 50
print("-" * tamanho)
```

```python
def imprimir_linha(tamanho: int = 50) -> None:
    '''
        Imprime uma linha de caracteres '-' com o tamanho especificado.
        Args:
            tamanho (int): O número de caracteres '-' a serem impressos.
        Returns:
            None
    '''
    print("-" * tamanho)
```

```python
# Chamando a função para imprimir uma linha de 50 caracteres
imprimir_linha()
```

```python
# Chamando a função para imprimir uma linha com o número de caracteres igual ao comprimento da string 'm'
imprimir_linha(len(m))
print(m)
imprimir_linha(len(m))
```

```python
# Chamando a função para imprimir uma linha com o número de caracteres igual ao comprimento da nova string 'm'
m = "Python delicinha"
imprimir_linha(len(m))
print(m)
imprimir_linha(len(m))
```

```python
def imprimir_mensagem(mensagem: str) -> None:
    '''
        Imprime uma mensagem centralizada entre linhas de caracteres '-'.
        Args:
            mensagem (str): A mensagem a ser impressa.
        Returns:
            None
    '''
    m = f"| {mensagem} |"
    imprimir_linha(tamanho=len(m))
    print(m)
    imprimir_linha(tamanho=len(m))
```

```python
# Chamando a função para imprimir uma mensagem personalizada
imprimir_mensagem(mensagem="Python é FODA")
```
