# Operadores Lógicos

## Simples

```python
# Verificando operadores lógicos entre inteiros
a = 4
b = 2

print(f"a == a: {(a == a)}")
print(f"b == b: {(b == b)}")
print(f"a != a: {(a != a)}")
print(f"b != b: {(b != b)}")
```

## Numpy

```python
import numpy as np
```

```python
# Verificando operadores lógicos entre arrays
a = np.array(('a', 'b', 'c', 'd', 'e'))
b = np.array((1, 2, 3, 4, 5))
print(f"a == b: {(a == b)}")
print(f"a == a: {(a == a)}")
print(f"b == b: {(b == b)}")
```

```python
# Verificando operadores lógicos entre arrays
a = np.array((6, 7, 8, 9, 10))
b = np.array((1, 2, 3, 4, 5))
print(f"a == b: {(a == b)}")
print(f"a == a: {(a == a)}")
print(f"b == b: {(b == b)}")
```

```python
# Verificando operadores lógicos entre arrays
try:
    print(f"{(a == a) and (b == b)}")
except Exception as e:
    print(f"Erro: {e}")
```

```python
# Verificando operadores lógicos entre arrays
print(np.all((a == a, b == b)))
print(np.all((a != a, b == b)))
print(np.all((a == a, b != b)))
print(np.all((a != a, b != b)))
```

```python
# Verificando operadores lógicos entre arrays
print(np.any((a == a, b == b)))
print(np.any((a != a, b == b)))
print(np.any((a == a, b != b)))
print(np.any((a != a, b != b)))
```

```python
# Verificando operadores lógicos entre arrays
a = np.array([1, 2, 3])
b = np.array([1, 2, 3_000])
print(f"a == b: {(a == b)}")
```

```python
# Verificando operadores lógicos entre arrays
c = np.array([4, 5, 6])

display(np.all((a == b, c == c)))
display(np.any((a == b, c == c)))
```
