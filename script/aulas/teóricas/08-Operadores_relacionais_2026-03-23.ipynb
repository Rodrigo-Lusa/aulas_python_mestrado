```python
# Verificando operadores relacionais entre inteiros
a = 2
b = 3
c = 4
print(f"{a} == {b}: {a == b}")
print(f"{a} != {b}: {a != b}")
print(f"{a} > {b}: {a > b}")
print(f"{a} >= {b}: {a > b}")
print(f"{a} < {b}: {a < b}")
print(f"{a} <= {b}: {a < b}")
print(f"{a} <= {b} < {c}: {a <= b < c}")
```

```python
# Verificando operadores relacionais entre inteiro e float
a = 2
b = 3.5
c = 4
print(f"{a} == {b}: {a == b}")
print(f"{a} != {b}: {a != b}")
print(f"{a} > {b}: {a > b}")
print(f"{a} >= {b}: {a > b}")
print(f"{a} < {b}: {a < b}")
print(f"{a} <= {b}: {a < b}")
print(f"{a} <= {b} < {c}: {a <= b < c}")
```

```python
# Verificando operadores relacionais entre inteiro e booleano
a = 1
b = False
c = 4
print(f"{a} == {b}: {a == b}")
print(f"{a} != {b}: {a != b}")
print(f"{a} > {b}: {a > b}")
print(f"{a} >= {b}: {a > b}")
print(f"{a} < {b}: {a < b}")
print(f"{a} <= {b}: {a < b}")
print(f"{a} <= {b} < {c}: {a <= b < c}")
```

```python
# Verificando operadores relacionais entre tuplas de inteiros e booleanos
obj_tupla_int = (1, 0, 1, 0)
obj_tupla_bool = (True, False, True, False)
print(f"{obj_tupla_int} == {obj_tupla_bool}: {obj_tupla_int == obj_tupla_bool}")
print(f"{obj_tupla_int} != {obj_tupla_bool}: {obj_tupla_int != obj_tupla_bool}")
print(f"{obj_tupla_int} > {obj_tupla_bool}: {obj_tupla_int > obj_tupla_bool}")
print(f"{obj_tupla_int} >= {obj_tupla_bool}: {obj_tupla_int > obj_tupla_bool}")
print(f"{obj_tupla_int} < {obj_tupla_bool}: {obj_tupla_int < obj_tupla_bool}")
print(f"{obj_tupla_int} <= {obj_tupla_bool}: {obj_tupla_int < obj_tupla_bool}")
```

```python
# Verificando operadores relacionais entre tupla de inteiros e lista de booleanos
try:
    obj_tupla_int = (1, 0, 1, 0)
    obj_list_bool = [True, False, True, False]
    print(f"{obj_tupla_int} == {obj_list_bool}: {obj_tupla_int == obj_list_bool}")
    print(f"{obj_tupla_int} != {obj_list_bool}: {obj_tupla_int != obj_list_bool}")
    print(f"{obj_tupla_int} > {obj_list_bool}: {obj_tupla_int > obj_list_bool}")
    print(f"{obj_tupla_int} >= {obj_list_bool}: {obj_tupla_int > obj_list_bool}")
    print(f"{obj_tupla_int} < {obj_list_bool}: {obj_tupla_int < obj_list_bool}")
    print(f"{obj_tupla_int} <= {obj_list_bool}: {obj_tupla_int < obj_list_bool}")
except Exception as e:
    print(f"Erro: {e}")
```

```python
# Verificando operadores relacionais entre listas de inteiro e booleano
obj_list_int = [1, 0, 1, 0]
obj_list_bool = [True, False, True, False]
print(f"{obj_list_int} == {obj_list_bool}: {obj_list_int == obj_list_bool}")
print(f"{obj_list_int} != {obj_list_bool}: {obj_list_int != obj_list_bool}")
print(f"{obj_list_int} > {obj_list_bool}: {obj_list_int > obj_list_bool}")
print(f"{obj_list_int} >= {obj_list_bool}: {obj_list_int > obj_list_bool}")
print(f"{obj_list_int} < {obj_list_bool}: {obj_list_int < obj_list_bool}")
print(f"{obj_list_int} <= {obj_list_bool}: {obj_list_int < obj_list_bool}")
```

```python
# Verificando operadores relacionais entre dicionários de inteiros e booleanos
try:
    obj_dict_int = {0: 1, 1: 0, 2: 1, 3: 0}
    obj_dict_bool = {0: True, 1: False, 2: True, 3: False}
    print(f"{obj_dict_int} == {obj_dict_bool}: {obj_dict_int == obj_dict_bool}")
    print(f"{obj_dict_int} != {obj_dict_bool}: {obj_dict_int != obj_dict_bool}")
    print(f"{obj_dict_int} > {obj_dict_bool}: {obj_dict_int > obj_dict_bool}")
    print(f"{obj_dict_int} >= {obj_dict_bool}: {obj_dict_int > obj_dict_bool}")
    print(f"{obj_dict_int} < {obj_dict_bool}: {obj_dict_int < obj_dict_bool}")
    print(f"{obj_dict_int} <= {obj_dict_bool}: {obj_dict_int < obj_dict_bool}")
except Exception as e:
    print(f"Erro: {e}")
```

```python
import pandas as pd
```

```python
# Verificando operadores relacionais entre pd.Series de inteiros e booleanos
obj_pandas_int = pd.Series([1, 0, 1, 0])
obj_pandas_bool = pd.Series([True, False, True, False])

print("-" * 40)
print(f"{obj_pandas_int == obj_pandas_bool}")
print("-" * 40)
print(f"{obj_pandas_int != obj_pandas_bool}")
print("-" * 40)
print(f"{obj_pandas_int > obj_pandas_bool}")
print("-" * 40)
print(f"{obj_pandas_int > obj_pandas_bool}")
print("-" * 40)
print(f"{obj_pandas_int < obj_pandas_bool}")
print("-" * 40)
print(f"{obj_pandas_int < obj_pandas_bool}")
```
