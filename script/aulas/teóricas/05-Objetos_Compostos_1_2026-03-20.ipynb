# Memória RAM

```python
lampada = "ligada"
# Obtém o ID (inteiro)
identificador = id(lampada)
print(f"Id: {identificador}") 

# Obtém o endereço em hexadecimal (formato de memória)
endereco_memoria = hex(id(lampada))
print(f"End hexa: {endereco_memoria}")
```
