```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Gerando os dados aleatórios com distribuição normal e uniforme
np.random.seed(42)
array_1 = np.random.normal(loc=0.0, scale=10.0, size=100)
array_2 = np.random.randint(0.0, 12.0, size=100)
array_3 = np.insert(np.array([array_1.mean()] * 5), 0, [np.nan]*95)
array_aula = list(array_1) + list(array_2) + list(array_3)
df = pd.DataFrame({"amostra": array_aula})
df = df.dropna()
df.info()
```

```python
# Criando o histograma com KDE e as linhas verticais para média, mediana e moda
plt.rcParams['lines.linewidth'] = 2.5
fig, ax = plt.subplots(figsize=(15, 5))
sns.histplot(
    df['amostra'],
    kde=True,
    bins=50,
    ax=ax
)
plt.axvline(df['amostra'].mean(), color='r', linestyle='--', label='Média')
plt.axvline(df['amostra'].median(), color='g', linestyle='--', label='Mediana')
plt.axvline(df['amostra'].mode()[0], color='y', linestyle='--', label='Moda')
plt.legend()
plt.title("Histograma com KDE")
plt.show()
```

```python
# Criando o boxplot com as linhas verticais para média, mediana e moda
fig, ax = plt.subplots(figsize=(15, 5))

sns.boxplot(
    x='amostra',
    ax=ax,
    notch=True,
    color='skyblue',
    data=df,
)

plt.axvline(df['amostra'].mean(), color='r', linestyle='--', label='Média')
plt.axvline(df['amostra'].median(), color='g', linestyle='--', label='Mediana')
plt.axvline(df['amostra'].mode()[0], color='y', linestyle='--', label='Moda')
plt.legend()
plt.title("Boxplot")
plt.xlabel("Amostra")
plt.ylabel("Valor")
plt.show()
```
