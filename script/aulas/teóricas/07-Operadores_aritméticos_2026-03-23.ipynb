```python
# Instanciando as variáveis inteiras
prim_num = 10
segn_num = 3
result = 0
```

```python
# Instanciando a tupla com as operações matemáticas usando as variáveis anteriores
tupla_opera = (
    prim_num + segn_num,
    prim_num - segn_num,
    prim_num * segn_num,
    round(prim_num / segn_num, 2),
    prim_num // segn_num,
    prim_num % segn_num,
    prim_num ** segn_num,
)
tupla_opera
```

```python
# Verificando se é possível usar um operador inteiro com a tupla
try:
    tupla_opera += 2
    print(tupla_opera)
except Exception as e:
    print(f"Error: {e}")
```

```python
# Anterando a tupla, com ela mesmo, concatenando
tupla_opera += tupla_opera
tupla_opera
```

```python
# Alterando a tupla, com ela mesma, multiplicando por um inteiro
tupla_opera *= 2
print(tupla_opera)
```

```python
import numpy as np
```

```python
# Instanciando um array do numpy com as mesmas operações matemáticas usando as variáveis anteriores
array = np.array((
    prim_num + segn_num,
    prim_num - segn_num,
    prim_num * segn_num,
    round(prim_num / segn_num, 2),
    prim_num // segn_num,
    prim_num % segn_num,
    prim_num ** segn_num,
))
print(array)
```

```python
# Modificando o array usando um operador inteiro
array += 2
print(array)
```

```python
# Modificando o array, com ele mesmo, somando valores
array += array
print(array)
```

```python
# Concatenando o array com ele mesmo
array_array = np.concatenate((array, array))
print(array_array)
```
