# Dicionários multi classe

```python
# Instanciando um dicionário para armazenar informações sobre a iluminação de um ambiente
locais = ['sala', 'cozinha', 'quarto', 'banheiro']
lampadas = ['ligada', 'desligada', 'quebrada', 'ligada']
num_lampdas = [2, 1, 1, 3]
testadas = [True, False, False, True]
metros_quadrados = [20.0, 15.2, 10.4, 5.9]

iluminação_1 = {
    'locais': locais,
    'lampadas': lampadas,
    'num_lampdas': num_lampdas,
    'testadas': testadas,
    'metros_quadrados': metros_quadrados
}
iluminação_1
```

```python
# Instanciando um dicionário para armazenar informações sobre a iluminação de um ambiente, utilizando tuplas e conjuntos
locais = ('sala', 'cozinha', 'quarto', 'banheiro')
lampadas = ['ligada', 'desligada', 'quebrada', 'ligada']
num_lampdas = (2, 1, 1, 3)
testadas = {
            locais[0]: True, 
            locais[1]: False, 
            locais[2]: False, 
            locais[3]: True
}
metros_quadrados = {20.0, 15.2, 10.4, 5.9}

iluminação_2 = {
    'locais': locais,
    'lampadas': lampadas,
    'num_lampdas': num_lampdas,
    'testadas': testadas,
    'metros_quadrados': metros_quadrados
}
iluminação_2
```

```python
# Verificando os tipos de dados armazenados em cada chave do dicionário iluminação_2
print(type(iluminação_2['locais']))
print(type(iluminação_2['lampadas']))
print(type(iluminação_2['num_lampdas']))
print(type(iluminação_2['testadas']))
print(type(iluminação_2['metros_quadrados']))
```

# Array

```python
# Alterando o dicionário para um array numpy
id = 'testadas'
array = np.array(iluminação_2['locais'])
array = np.append(array, iluminação_2['lampadas'])
array = np.append(array, iluminação_2['num_lampdas'])
array = np.append(array, [
                    iluminação_2[id][list(iluminação_2[id].keys())[0]],
                    iluminação_2[id][list(iluminação_2[id].keys())[1]],
                    iluminação_2[id][list(iluminação_2[id].keys())[2]],
                    iluminação_2[id][list(iluminação_2[id].keys())[3]]
                ]
)
array = array.reshape(4, -1)
array
```
