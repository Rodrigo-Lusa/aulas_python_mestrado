```python
# Usando um loop while para preencher um vetor lista
vetor_num = []
contador = 10
while contador <= 60:
    vetor_num.append(contador)
    contador += 10
vetor_num
```

```python
# Usando um loop for para preencher um vetor lista

vetor_num = []
contador = 10
for _ in range(6):
    vetor_num.append(contador)
    contador += 10
vetor_num
```
